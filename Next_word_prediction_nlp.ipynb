```python
import pandas as pd
import numpy as np
import re
import nltk
```

```python
dataset = pd.read_csv("Restaurant_Reviews.tsv", delimiter="\t")
```

```python
dataset.drop("Liked", axis=1, inplace=True)
```

```python
dataset
```

```python
corpus=[]
for i in range(1000):
    review=dataset["Review"][i]
    review = re.sub("[^a-zA-Z]"," ",review)
    review = review.lower()
    corpus.append(review)
```

```python
corpus
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(corpus)
```

```python
tokenizer.word_index
```

```python
token_list = tokenizer.texts_to_sequences(corpus)
```

```python
token_list
```

```python
input_sequences=[]
```

```python
for review in corpus:
    token_list = tokenizer.texts_to_sequences([review])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
max_sequence_len = max([len(x) for x in input_sequences])
```

```python
max_sequence_len
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
input_sequences=pad_sequences(input_sequences,max_sequence_len,padding="pre")
```

```python
input_sequences.shape
```

```python
x=input_sequences[:,:-1]
```

```python
y=input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
import keras
```

```python
y=keras.utils.to_categorical(y,num_classes=2023)
```

```python
y
```

```python
from keras.layers import Embedding
```

```python
from keras.models import Sequential
```


# Init NN model

```python
total_words=2023
```

```python
model=Sequential()
```

```python
model.add(Embedding(input_dim=total_words,output_dim=15,input_length=max_sequence_len-1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

# INIT NN model

```python
model1=Sequential()
```

```python
from keras.layers import LSTM
```

```python
from keras.layers import Dense
```

```python
total_words=2023
```

```python
model1.add(Embedding(input_dim=total_words,output_dim=15,input_length=31))
```

```python
model1.add(LSTM(return_sequences=True,units=200,kernel_initializer='glorot_uniform'))
```

```python
model1.add(LSTM(return_sequences=True,units=100,kernel_initializer='glorot_uniform'))
```

```python
model1.add(LSTM(return_sequences=False,units=100,kernel_initializer='glorot_uniform'))
```


# output layers

```python
model1.add(Dense(kernel_initializer='glorot_uniform',activation='softmax',units=total_words))
```

```python
model1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model1.summary()
```

```python
# train and test divide based on iloc like time series
```

```python
model1.fit(x,y,epochs=200,batch_size=32)
```
